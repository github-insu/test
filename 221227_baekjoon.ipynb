```python
T = int(input())
result = 0
for i in range(T):
    A,B = map(int, input().split())
    gob = A*B
    multi = 1
    count = 2
    while True:
        if (A % count == 0) and (B % count == 0):
            multi *= count
            A /= count
            B /= count
        else :
            if count > min(A, B) :
                break
            count += 1
    result = int(gob/multi)
    print(result)
```
